```python
numeric_features=[feature for feature in data.columns if data[feature].dtype !='0']
```

```python
numeric_features
```

```python
plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of Numerical Features',fontsize=20,fontweight='bold')
for i in range(0,len(numeric_features)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=data[numeric_features[i]],fill=True,color='r')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```

```python
x_axis=numeric_features
y="Temperature"
for col in x_axis:
    sns.regplot(x=col,y=y,data=data)
    plt.xlabel(col,fontsize=13)
    plt.ylabel(y,fontsize=13)
    plt.title("Scatter plot of "+ col +" & Temperature")
    plt.grid()
    plt.show()
    
```
